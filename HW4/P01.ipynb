```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import time

plt.rcParams["figure.figsize"] = (12,6)
```

```python
img_1 = cv2.imread("./img1.jpg",0)
img_2 = cv2.imread("./img2.jpg",0)

def KMeansSeg(img:np.ndarray):
    img = img.copy()
    h,w = img.shape
    img_arr = img.reshape(h*w,1)
    scaler = StandardScaler()
    scaler.fit(img_arr)
    img_scl = scaler.transform(img_arr)

    start = time.time()
    kmeans = KMeans(3)
    kmeans.fit(img_scl)
    end = time.time()
    pred = kmeans.labels_

    segmented = pred.reshape((h,w))
    plt.imshow(segmented)
    plt.axis("off")
    dur = round(end-start,2)
    plt.title("Segmentation using K-Means with Normalization | Time:{0}".format(dur))
    plt.show()
```

```python
KMeansSeg(img_1)
```

```python
KMeansSeg(img_2)
```
